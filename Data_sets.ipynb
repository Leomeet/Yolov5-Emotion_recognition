```python
'''

train: ../train/images
val: ../test/images

nc: 7
names: ['angry', 'disgust', 'fear', 'happy', 'neutral', 'sad', 'surprise']
'''
# angry:0 disgust:1 fear:2 happy:3 neutral:4 sad:5 surprise:6
import os
main = r"M:\_02_ # Github Clones\Yolov5-Emotion_recognition\FUR_2013\test\images"#folder path
sub_folder = "surprise"
landing_dir = r"M:\_02_ # Github Clones\Yolov5-Emotion_recognition\FUR_2013\test\lables"#folder path for lables
coun = 0

for file_name in os.listdir(main):
    if file_name.endswith(".jpg"):
        txt_name = os.path.splitext(file_name)[0]
        txt_name = txt_name+".txt"
        lable_file = os.path.join(landing_dir,txt_name)
        file = open(lable_file,'w')
        if "angry" in lable_file:
            file.write("0 0.5 0.5 1 1")
        elif "disgust" in lable_file:
            file.write("1 0.5 0.5 1 1")
        elif "fear" in lable_file:
            file.write("2 0.5 0.5 1 1")
        elif "happy" in lable_file:
            file.write("3 0.5 0.5 1 1")
        elif "neutral" in lable_file:
            file.write("4 0.5 0.5 1 1")
        elif "sad" in lable_file:
            file.write("5 0.5 0.5 1 1")
        elif "surprise" in lable_file:
            file.write("6 0.5 0.5 1 1")
            
        file.close()
    coun+=1
    print(file_name)
    
print(coun)
```
